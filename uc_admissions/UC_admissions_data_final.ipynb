```python
uc.head()
```

```python
uc_admits = uc[uc['Category'] == 'Adm']
```

```python
uc_apps = uc[uc['Category'] == 'App']
```

```python
uc_apps
```

```python
merged_uc = pd.merge(uc_admits, uc_apps, on=['Calculation1', 'School', 'City', 'County/State/ Territory', 'Ethnicity'])
```

```python
merged_uc.head()
```

```python
merged_uc.rename(columns={'Count_x' : 'Count_adm', 'Count_y' : 'Count_app'}, inplace=True)
```

```python
merged_uc.drop(['Category_x', 'Category_y'], axis=1, inplace=True)
```

```python
merged_uc['admit_rate'] = (merged_uc['Count_adm'] / merged_uc['Count_app'])*100
```

```python
x = merged_uc.groupby('Ethnicity')['Count_app'].sum()
```

```python
overall_admit_ethn = pd.merge(x, y, on='Ethnicity')
```
